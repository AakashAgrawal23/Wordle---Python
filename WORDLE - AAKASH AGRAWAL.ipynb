```python
'''FUNCTION FOR ACCEPTING COLOR CODE OF THE GUESSED LETTER BY USER'''

def color():

    while True:

        color = input('Please enter your color code: Y or G or B ')

        if color.lower() == 'y' or color.lower() == 'g' or color.lower() == 'b':
            break
        else:
            print('Please follow the instructions!')
    
    return color

'''FUNCTION FOR ACCEPTING THE GUESSED LETTER BY USER'''

def letter():
    
    while True:
        
            letter = input('Please enter your guess here: (One letter only) ')
        
            if letter.isalpha() and len(letter) == 1:
                break
            else:
                print('Please follow the instructions!')
    
    return letter


from IPython.display import clear_output
from os import system

dict1 = {}
'''CREATED DICTIONARY TO STORE LETTER AND ITS COLOR CODE ALONG IT'S POSITION'''

temp_list = wordlist
'''DONT WANT TO MESS WITH MY MAIN WORDLIST, HENCE MADE A DUMMY TEMP_LIST'''

waste_list = []
'''WASTE LIST WILL CONTAIN THE WORDS THAT ARE CANCELLED OUT BY THE USERS GUESSES'''

trial = 0
'''TRIAL WILL COUNT THE NUMBER OF ATTEMPTS MADE BY USER'''

game_on = True
'''WHILE THE GAME IS ON... THE CODE RUNS'''

while game_on:      #The loop starts with game_on being TRUE
    
    clear_output()  #To clear the screen, specific to Jupyter Notebook
    
    trial += 1      #Trial count increased by one, counting one attempt
    print(f'Trial ----- {trial} guesses:\n')
    x = len(temp_list)
    print(f'Your probability of getting right result is 1 in {x}\n')
    
    
    '''THE FOR LOOP TO ASK USER TO ENTER THE GUESSED LETTER AND COLOR CODE 5 TIMES'''
    
    for box in range(5):
        
        print(f'Now enter your guess letter for position {box+1}: \n')
        
        letter_guess = letter()
                
        color_guess = color()
        
        temp_dict = {box : [letter_guess.lower(),color_guess.lower()]}
        
        dict1.update(temp_dict)
        
    '''HEART AND SOUL OF THE CODE.... USING THE DICTIONARY CREATED ABOVE AND BASED ON THE CONDITIONS
    OF THE COLOR CODE OF THE LETTER, CALCULATES THE NUMBER OF WORDS WITH PROBABLE MATCHES!!!!
    
    IF THE CODE IS B AS IN BLACK... THE LETTER IS NOT IN THE WORD
    
    IF THE CODE IS Y AS IN YELLOW.... THE LETTER EXISTS IN THE WORD BUT NOT AT THAT LOCATION
    
    IF THE CODE IS G AS IN GREEN.... THE LETTER EXISTS IN THE WORD AND IS AT THAT LOCATION
    
    USED WASTE LIST AS A DUMMY LIST TO APPEND WORDS THAT WE DONT WANT
    
    USED LIST COMPREHENSION TO REMOVE WORDS FROM THE TEMP_LIST THAT EXISTS IN WASTE_LIST
    
    THE UPDATED WASTE_LIST IS USED IN CONSEQUENT LOOP'''
    
    for box in range(5):
        
        if dict1[box][1] == 'b':
            
            for word in temp_list:
                
                if dict1[box][0] in word:
                    waste_list.append(word)
                    
        elif dict1[box][1] == 'y':
            
            for word in temp_list:
                
                if dict1[box][0] == word[box]:
                    waste_list.append(word)
        
        else:
            
            for word in temp_list:
                
                if dict1[box][0] != word[box]:
                    waste_list.append(word)
    
    temp_list = [i for i in temp_list if i not in waste_list]
    
    
    '''STATS CALCULATED TO BE DISPLAYED TO THE USER'''
    y = len(temp_list)
    p_x = 1/x
    p_y = 1/y
    p_diff = (p_y-p_x)*100/p_x

    if y == 1: #IF THE LEN(TEMP_LIST) IS 1 MEANS, THAT IS THE WORD ITSELF, THE USER WON
        print(f'\nCongratulations!!!!!!!!!!!! YOU DID IT IN {trial} try! IMPRESSIVE \n')
        game_on = False
        break
    else: #USER STILL HAS NOT BEEN ABLE TO GUESS THE CORRECT WORD
        print(f'The correct answer is one of the {y} words\n')
        print(f'That is the probability increase of {p_diff} %') 

    if(trial<7):
        play = input('Do you want to continue?: Y or N ')

        if play.lower() == 'y':
            game_on = True
        else:
            game_on = False

print('\n\tI hope you enjoyed being nerd! You are SMARTER than you think')
```
